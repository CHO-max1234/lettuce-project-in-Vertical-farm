```python
import os

image_root = "/content/drive/Othercomputers/내 컴퓨터/새 폴더/양상추사진/양상추_테라웨이브/양상추 날짜/RGB_정면/4. 결과 출력 시각화/2작기 error/260421 piece crop_ex/VER2_Gemini/crop"

# EasyOCR 리더 초기화 (한국어와 영어를 지원하도록 설정)
reader = easyocr.Reader(['ko', 'en'])

# 이미지 파일 목록 가져오기
image_files = [os.path.join(image_root, f) for f in os.listdir(image_root) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
image_files.sort() # 파일 순서를 일관성 있게 정렬

print(f"Processing first {min(5, len(image_files))} images from {image_root}")

# 첫 5개의 이미지에 대해 EasyOCR 수행
for i, image_path in enumerate(image_files[:5]):
    print(f"\n--- Processing image {i+1}: {os.path.basename(image_path)} ---")
    img = cv2.imread(image_path)

    if img is None:
        print(f"Error: Could not read image {image_path}")
        continue

    # 이미지에서 텍스트 인식
    results = reader.readtext(img)

    # 결과 출력
    if results:
        for (bbox, text, prob) in results:
            print(f"Detected: {text}, Confidence: {prob:.4f}")
    else:
        print("No text detected.")
```

```python
def get_bed_number(img_path):
    # [포인트 1] 설정값(config)
    # --psm 6: "이 사진은 줄글이 아니라 하나의 덩어리(블록)야"라고 알려주는 것
    # whitelist: "알파벳이나 한글은 무시하고 0~9 숫자만 찾아!"라고 명령하는 필터
    config = '--psm 6 -c tessedit_char_whitelist=0123456789'

    # [포인트 2] 변환 함수(image_to_string)
    # 이미지를 넣으면 컴퓨터가 읽을 수 있는 '텍스트(문자열)'로 뱉어내는 핵심 기능
    result = pytesseract.image_to_string(img_path, config=config)

    # [포인트 3] strip()
    # 글자 앞뒤에 붙은 불필요한 공백이나 줄바꿈을 지워주는 깔끔이 작업
    return result.strip()
```

```python
import pytesseract
import cv2

# 코랩 환경용 Tesseract 경로 설정
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

image_path = "/content/drive/Othercomputers/내 컴퓨터/새 폴더/양상추사진/양상추_테라웨이브/양상추 날짜/RGB_정면/4. 결과 출력 시각화/2작기 error/260421 piece crop_ex/VER2_Gemini/crop/B/bed01_20260318_131655_cam1_piece.jpg"

def get_bed_number(img_path):
    """
    이미지를 읽어 OCR로 숫자를 추출하는 함수
    """
    img = cv2.imread(img_path)

    if img is None:
        print(f"파일을 찾을 수 없습니다: {img_path}")
        return None

    # 2. OCR 설정 (숫자만 추출)
    custom_config = r'--psm 8 -c tessedit_char_whitelist=0123456789'

    # 3. 추출
    result = pytesseract.image_to_string(img, config=custom_config)

    return result.strip()

# 함수를 호출하여 결과만 출력합니다.
result_value = get_bed_number(image_path)
print(f"인식된 숫자: {result_value}")
```

```python
import pytesseract
import cv2
import os
image_path = "/content/drive/Othercomputers/내 컴퓨터/새 폴더/양상추사진/양상추_테라웨이브/양상추 날짜/RGB_정면/4. 결과 출력 시각화/2작기 error/260421 piece crop_ex/VER2_Gemini/crop/B/bed01_20260318_131655_cam1_piece.jpg"
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

def get_bed_number_for_white_bg(img_path):
    if not os.path.exists(img_path):
        return "Path Error"

    # 1. 이미지 로드 (흑백 모드)
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

    # 2. [추가 권장] 아주 살짝만 안쪽으로 크롭 (테두리 노이즈 제거)
    # 이미지 가장자리의 검은 선들을 5% 정도 깎아냅니다.
    h, w = img.shape
    margin = 0.05
    img = img[int(h*margin):int(h*(1-margin)), int(w*margin):int(w*(1-margin))]

    # 3. OCR 설정
    # --psm 8: 이미지에 단어 하나만 있다고 가정 (가장 강력함)
    # whitelist: 숫자만 나오도록 제한
    config = r'--psm 8 -c tessedit_char_whitelist=0123456789'

    # 4. 인식 (반전 없이 그대로 진행)
    result = pytesseract.image_to_string(img, config=config)

    return result.strip()

# 실제 경로로 테스트
print(f"결과: {get_bed_number_for_white_bg(image_path)}")
```

cam1 원본

-> EasyOCR Reader 1회 로드 후 재사용
-> allowlist 숫자 OCR
-> 1~95 범위 체크
-> confidence 낮으면 이전/다음 bed 흐름으로 QC
-> real_bed_id 확정
그리고 매 프레임 OCR할 필요도 없다. 촬영 triplet당 한 번, 또는 bed가 바뀌는 시점마다 한 번이면 된다. reader를 매번 새로 만들지만 않으면 모델 4개라는 부담도 꽤 줄어든다.

결론은 이거야. 정확성이 제일 중요하면 EasyOCR 메인으로 가는 게 맞다. Tesseract는 “가벼운 대신 앞뒤 코드가 무거워지는 선택지”가 될 가능성이 크다. 지금 이 문제는 bed identity가 틀리면 아래 모델이 다 무너지는 상류 문제라서, 여기에는 조금 무거워도 한 방 정확도가 높은 도구를 쓰는 게 맞다.
